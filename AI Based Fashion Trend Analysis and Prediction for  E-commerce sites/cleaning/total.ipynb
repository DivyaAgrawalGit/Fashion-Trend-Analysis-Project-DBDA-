```python
import pandas as pd
import glob
import json

def merge_json_files_pandas(file_list):
    combined_data = pd.DataFrame()
    for json_file in file_list:
        with open(json_file, 'r') as infile:
            file_data = json.load(infile)
            df = pd.DataFrame(file_data)
            combined_data = pd.concat([combined_data, df], ignore_index=True)
    return combined_data
    

# Example usage
files_to_merge = glob.glob('*.json')  # Get all JSON files in the current directory
df=merge_json_files_pandas(files_to_merge)
```

combined_data.to_json('merged_file_pandas.json', orient='records')

```python
persons=['Men','Women','Kid','Girl','Boy']
def idealfor(row):
    z=row['Title']
    for x in persons:
        if x in z:   # Make case-insensitive
            return x 
        elif 'Lehenga' in z:
            return 'Women'
        elif 'Saree' in z:
            return 'Women'
        elif 'Trouser' in z:
            return 'Men'
        elif 'Shirt' in z:
            return 'Men'
        elif 'Kurta' in z:
            return 'Women'
        elif 'Salwar' in z:
            return 'Men'
```

```python
df['Ideal for']=df.apply(idealfor,axis=1)
df['Ideal for'].value_counts()
```
