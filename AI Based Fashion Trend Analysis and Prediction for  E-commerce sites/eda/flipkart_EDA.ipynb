```python

import matplotlib.pyplot as plt

# Assuming df is your DataFrame
df_blue = df[df['Color'] == 'Blue']  # Filter out rows where 'Color' is not 'Blue'

# Sum the counts of different star ratings for the 'Color' column with value 'Blue'
star_counts = df_blue[['1 star', '2 stars', '3 stars', '4 stars', '5 stars']].sum()

# Create a single pie chart for the star counts
plt.pie(star_counts, labels=star_counts.index, autopct='%1.1f%%')
plt.title('Overall distribution of charts for Blue color')
plt.show()
```

```python
# Assuming df is your DataFrame
top_colors = df['Color'].value_counts().nlargest(5).index.tolist()
top_ratings = df['Rating'].value_counts().nlargest(5).index.tolist()

# Filter the DataFrame based on the top colors and top ratings
df_filtered = df[df['Color'].isin(top_colors) & df['Rating'].isin(top_ratings)]

# Group the filtered DataFrame by 'Color' and 'Rating' and count the number of occurrences
counts = df_filtered.groupby(['Color', 'Rating']).size().unstack()

# Create a stacked bar chart for the top 5 colors and their ratings
fig, ax = plt.subplots()
counts.plot(kind='bar', stacked=True, ax=ax)
ax.set_xlabel('Color')
ax.set_ylabel('Count')
ax.set_title('Stacked bar chart for top 5 colors and their ratings')
plt.show()
```
