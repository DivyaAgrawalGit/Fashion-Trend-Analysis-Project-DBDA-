```python
import string
import random
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
def comments(url):
    print(url)
    #time.sleep(0.5)
    r= requests.get(url)
    #print (r)
    soup = BeautifulSoup(r.text, "html.parser")
    #return soup
    reviews=soup.find("div", {'class':"col JOpGWq _33R3aa"})
    dis=soup.find('div',{'class':'_30jeq3 _16Jk6d'})
    #print(dis)
    if dis:
        dis=dis.text.strip()
    if reviews:
        reviews=reviews.findAll('a')
        if reviews:
            reviews=reviews[-1]
            all_reviews='https://www.flipkart.com'+reviews['href']
        else:
            all_reviews=None
    else:
        all_reviews=None
    
    dark=soup.findAll('div',{'class':'col col-3-12 _2H87wv'})
    light=soup.findAll('div',{'class':'col col-9-12 _2vZqPX'})
    des={}
    star=[0,0,0,0,0]
    for i in range(len(light)):
        des[dark[i].text.strip()]=light[i].text.strip()
    code=generate_unique_code()
    if reviews:
        #print(all_reviews)
        r= requests.get(all_reviews)
        #print(r)
        soup = BeautifulSoup(r.text, "html.parser")
        #print(soup)
        stars=soup.findAll('div',{'class':'_1uJVNT'})
        #print(stars)
        
        for i in range(len(stars)):
            star[i]=stars[4-i]
            if star[i]:
                star[i]=star[i].text.strip()
            else:
                star[i]=0
    return [dis,code,des,star,all_reviews] 



def generate_unique_code():
    characters = string.ascii_letters + string.digits  # All possible characters
    unique_code = ''.join(random.sample(characters, 10))  # Randomly sample 10 characters without replacement
    return unique_code
```

```python
df=pd.read_csv('indi_club.csv')
df['total']=df.apply(lambda x:comments(x.Url),axis=1)
df.to_json('indi_club.json')
```
